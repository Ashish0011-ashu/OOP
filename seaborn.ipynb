# Dist Plot
plots with Bar Graph and Kernel Density Estimate

```python
import seaborn as sns
```

```python
sns.distplot(df['average_monthly_hours'])
```

```python
sns.displot(df['number_projects'])
```

```python
# Histograms represent the data distribution by
# forming bins along the range of the data and then
# draing bars to show the numbers of obervations
# that fall in each  bin.
```

```python
df.describe()
```

```python
bins =  [2,3,4,5,6,8,9,10]
sns.distplot(df['time_spend_company'],bins= bins)
plt.xticks(bins)
```

```python
bins =  [2,3,4,5,6,8,9,10]
sns.distplot(df['number_projects'],bins= bins , rug = True , hist_kws= {'color':'red' , 'edgecolor':'blue' , 'linewidth':3,'alpha':1},
             kde_kws= {'color':'black' ,  'linewidth':3,'alpha':1})
plt.xticks(bins)  # alpha is for intensity of color
```

# Scatter plot

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.scatterplot(x = 'age' , y = 'fare' ,
                data = titanic_df)
```

```python
sns.scatterplot(x = 'age' , y = 'fare' ,
                data = titanic_df , hue = 'sex')
```

```python
sns.scatterplot(x = 'age' , y = 'fare' ,
                data = titanic_df , hue = 'alive')
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = 'age' , y = 'fare' ,
                data = titanic_df , hue = 'alive' ,  style = 'class')
plt.title('titanic data analysis')
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = 'age' , y = 'fare' ,
                data = titanic_df , hue = 'alive' ,  style = 'class' , s = 100 , palette = 'gist_rainbow')
plt.title('titanic data analysis')
```

```python
plt.figure(figsize = (12,7))
sns.scatterplot(x = 'age' , y = 'fare' ,
                data=titanic_df , hue = 'alive' ,  style = 'class' , s = 100 , palette = 'gist_rainbow' , alpha = 0.5)
plt.title('titanic data analysis')
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age' , y = 'fare' , data = titanic_df,
style = 'class')
sns.lineplot(x = 'age' , y= 'fare' , data = titanic_df , color = 'red')
plt.title('titanic data analysis')
plt.show()
```

# BAR PLOT

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.barplot(x = 'sex' , y = 'age' , data = titanic_df)
```

```python
sns.barplot(x = 'sex' , y = 'age' ,  data = titanic_df , hue = 'alive' )
```

# A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
sns.barplot(x = 'sex' , y = 'age' ,  data = titanic_df ,
            hue = 'fare' , ci = None , palette= 'icefire')
```

```python
sns.barplot(y = 'class' , x = 'fare' , data = titanic_df , hue = 'sex'
            , palette= 'icefire' , orient = 'h')
```

```python
sns.barplot(x = 'class' , y = 'fare' ,  data = titanic_df ,
            hue = 'sex' ,   palette= 'inferno' , estimator = np.median)

# estimator is for mean ,median of the bar data
```

```python
sns.barplot(x = 'class' , y = 'fare' ,  data = titanic_df ,
            hue = 'sex' , ci = 100 , palette= 'icefire' , errcolor='red' ,errwidth= 10)
# ci varies between 0 to 100 for that black line
# errcolor and errwidth decide the color and width of that line on bar
```

```python
sns.barplot(x = 'class' , y = 'fare' ,  data = titanic_df ,
            hue = 'sex' , saturation=0.3, palette= 'icefire')
```
