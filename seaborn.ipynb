# Dist Plot
plots with Bar Graph and Kernel Density Estimate

```python
import seaborn as sns
```

```python
sns.distplot(df['average_monthly_hours'])
```

```python
sns.displot(df['number_projects'])
```

```python
# Histograms represent the data distribution by
# forming bins along the range of the data and then
# draing bars to show the numbers of obervations
# that fall in each  bin.
```

```python
df.describe()
```

```python
bins =  [2,3,4,5,6,8,9,10]
sns.distplot(df['time_spend_company'],bins= bins)
plt.xticks(bins)
```

```python
bins =  [2,3,4,5,6,8,9,10]
sns.distplot(df['number_projects'],bins= bins , rug = True , hist_kws= {'color':'red' , 'edgecolor':'blue' , 'linewidth':3,'alpha':1},
             kde_kws= {'color':'black' ,  'linewidth':3,'alpha':1})
plt.xticks(bins)  # alpha is for intensity of color
```
