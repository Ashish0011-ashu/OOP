# # Merge() Function

```python
df1 = pd.DataFrame({'ROLL_NO' : [1,2,3,4,5] , 'PHYSICS' :[53,54,56,52,85]})
df1
```

```python
df2 = pd.DataFrame({'ROLL_NO' : [1,2,3,4,5] , 'MATHMATICS' :[20,54,32,42,85]})
df2
```

```python
# now merge two dataframe
pd.merge(df1,df2 , on = 'ROLL_NO')
```

```python
pd.merge(df2,df1 , on = 'ROLL_NO')
```

```python
df3  = pd.DataFrame({'ROLL_NO' : [1,2,7,9,5] , 'PHYSICS' :[53,54,56,52,85]})
df4 = pd.DataFrame({'ROLL_NO' : [1,2,6,8,5] , 'MATHMATICS' :[20,54,32,42,85]})
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4 , how='left') #only consisder df3 roll no
```

```python
pd.merge(df3 , df4 , how='right')  #only consisder df4 roll no
```
