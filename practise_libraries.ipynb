```python
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# function to plot pmf and pdf
def plot_discrete_dist(dist_name,x_values,pmf_values,cdf_values):
  plt.figure(figsize = (14,6))

  # plot probability mass function(pmf)
  plt.subplot(1,2,1)
  plt.stem(x_values , pmf_values)
  plt.title(f"pmf plot of{dist_name} distribution")
  plt.xlabel("k")
  plt.ylabel("p(x=k)")
  plt.grid(True)

  # plot cdf
  plt.subplot(1,2,2)
  plt.stem(x_values , cdf_values)
  plt.title(f"cdf plot of{dist_name} distribution")
  plt.xlabel("k")
  plt.ylabel("p(x<=k)")
  plt.grid(True)
```

```python
import numpy as np
from scipy.stats import binom

n =  int(input("enter the number of trials:"))
p = 0.5
x_binom = np.arange(0,n+1)
pmf_binom = binom.pmf(x_binom,n,p)
cdf_binom = binom.cdf(x_binom,n,p)
dist_name = "binomial"
plot_discrete_dist(dist_name,x_binom,pmf_binom,cdf_binom)
```

```python
# Binomial distribution without using scipy
from math import comb
def binomial_dist(n,p):
  x =np.arange(0,n+1)
  pmf = [comb(n,k)*(p**k)*((1-p)**(n-k)) for k in x]
  cdf = np.cumsum(pmf)
  return x , pmf , cdf

```

```python
n = int(input("enter the number of trails: "))
p =0.9
x_binom,pmf_binom,cdf_binom = binomial_dist(n,p)
dist_name = "binomial"
plot_discrete_dist(dist_name,x_binom,pmf_binom,cdf_binom)
```
