```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000) # Increased the range of t to 0-40
```

```python
d_r = 2.5*t   # distance travel by robber
```

```python
d_s = 3*(t-5)  # distance travel by shiff
```

```python
fig, ax = plt.subplots()
plt.title("A bank Robber caught")
plt.xlabel("time(in min)")
plt.ylabel("distance(in km)")
ax.set_xlim([0,40]) # Corrected typo from set_xlm to set_xlim
ax.set_ylim([0,400])
ax.plot(t,d_r , c = "green")
ax.plot(t,d_s , c = "brown")
plt.axvline(x =30 , color = "purple" , linestyle = '--') # Updated x-intercept to 30
plt.axhline(y = 75 , color = "purple" , linestyle = '--') # Updated y-intercept to 75
plt.show()
```
